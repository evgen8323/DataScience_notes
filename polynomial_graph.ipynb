```python
import numpy as np
import matplotlib.pyplot as plt

def f(A, x):
    y = np.array([0.0] * len(x))
    for i in range(len(A)):
        y += A[i] * x**(len(A) - 1 - i)
    return y

n = 6
X = np.random.randint(-n * 10, n * 10, n)
Y = np.random.randint(-n * 10, n * 10, n)

m = []
for i in range(n):
    mm = []
    for j in range(n - 1, -1, -1):
        mm.append(X[i] ** j)
    m.append(mm)
    
m = np.array(m)
y = np.array(Y)
m_inv = np.linalg.inv(m)
A = m_inv @ y

delta = (max(X) - min(X)) // 10
x = np.linspace(min(X) - delta, max(X) + delta)
y = f(A, x)

plt.figure(figsize=(16, 6))
_ = plt.scatter(X, Y, c='black')
_ = plt.plot(x, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(A, x):
    y = np.array([0.0] * len(x))
    for i in range(len(A)):
        y += A[i] * x**(len(A) - 1 - i)
    return y

n = 12

X = np.random.uniform(low=-n * n, high=n * n, size=(n,))
Y = np.random.uniform(low=-n * n, high=n * n, size=(n,))

m = []
for i in range(n):
    mm = []
    for j in range(n - 1, -1, -1):
        mm.append(X[i] ** j)
    m.append(mm)
    
m = np.array(m)
y = np.array(Y)
m_inv = np.linalg.inv(m)
A = m_inv @ y

x = np.linspace(min(X), max(X), n * 20)
y = f(A, x)

fig, axs = plt.subplots(2, 1, sharex=True)
fig.subplots_adjust(hspace=0.2)
fig.set_figheight(20)
fig.set_figwidth(30)

axs[0].scatter(X, Y, c='black')
axs[0].plot(x, y)

axs[1].scatter(X, Y, c='black')
axs[1].set_ylim(1.1 * min(Y), 1.1 * max(Y))
axs[1].plot(x, y)
```
